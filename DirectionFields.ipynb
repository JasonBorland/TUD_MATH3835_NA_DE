```python

## Library
import numpy as np
import math 

%matplotlib inline
import matplotlib.pyplot as plt # side-stepping mpl backend
import matplotlib.gridspec as gridspec # subplots
import warnings

warnings.filterwarnings("ignore")
```

```python
# Differential equation
# diff = y'= y/x (or say x+y)
def diff1(x,y):
    return x-y# 

def diff2(x,y):
    return 2*x + y# 

def diff3(x,y):
    return 1/(x+y)# 
```

```python
x = np.linspace(-2,2,10)
y = np.linspace(-2,2,10)

X, Y = np.meshgrid(x, y)


deg1 = np.arctan(diff1(X,Y))
dx1 = np.cos(deg1)
dy1 = np.sin(deg1)

deg2 = np.arctan(diff2(X,Y))
dx2 = np.cos(deg2)
dy2 = np.sin(deg2)


x3 = np.linspace(-2,2,15)
y3 = np.linspace(-2,2,15)
X3, Y3 = np.meshgrid(x3, y3)

deg3 = np.arctan(diff3(X3,Y3))
dx3 = np.cos(deg3)
dy3 = np.sin(deg3)
```

```python


fig, axs = plt.subplots(3,figsize=(10,30))
n = -1
color_array1 = np.sqrt(((dx1-n)/2)**2 + ((dy1-n)/2)**2)
color_array2 = np.sqrt(((dx2-n)/2)**2 + ((dy2-n)/2)**2)
color_array3 = np.sqrt(((dx3-n)/2)**2 + ((dy3-n)/2)**2)


axs[0].quiver(X,Y,dx1,dy1, color_array1)
axs[0].set_title('f(x,y) = x-y')

axs[1].quiver(X,Y,dx2,dy2, color_array2)
axs[1].set_title('f(x,y) = 2x + y')

axs[2].quiver(X3,Y3,dx3,dy3, color_array3)
axs[2].set_title('f(x,y) = 1/(x+y)')

for ax in axs.flat:
    ax.set(xlabel='x-label', ylabel='y-label')
    ax.grid(True)

#plt.show()
```

```python
fig = plt.figure(figsize=(10,10))
plt.quiver(X,Y,dx1,dy1, color_array1, scale=17)
plt.xlim((-2,2))
plt.ylim((-2,2))
plt.title('Illustration of discrete time points')
plt.grid(True)

plt.show()
```
