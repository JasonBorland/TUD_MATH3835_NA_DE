Find the stationary points of the following functions and determine there nature:

### Definition of a Relative Minimum
A function $f(x,y)$ has a relative minimum at a point (a,b) if $f(x,y) \ge f(a,b)$ for all point (x,y) in the same region of (a,b).

### Definition of a Relative Maximum
A function $f(x,y)$ has a relative maximum at a point (a,b) if $f(x,y) \le f(a,b)$ for all point (x,y) in the same region of (a,b).

### Definition of a Critical or Stationary point
The point (a,b) is a critical (or stationary) point of $f(x,y)$ provided one of the following is true:
1. $\nabla f(a,b) = \vec{0}$ i.e. $f_x(a,b) = 0 \ and \ f_y(a,b) = 0$ 
2. $f_x(a,b) \ and \ or \ f_y(a,b)$  does not exits.

### Theorem of critical points of a function $z = f(x,y)$

Let $z = f(x,y)$ be a function of two variables which has a stationary (critical) point at $(x_0,y_0)$.  This means 

$$\frac{\partial f}{\partial x} \Big \lvert _{x_0,y_0} = 0 . \ \ \frac{\partial f}{\partial y} \Big \lvert _{x_0,y_0} = 0.$$

If 
$$ 
D = 
\begin{vmatrix}
\frac{\partial ^2 f}{\partial x^2} & \frac{\partial ^2 f}{\partial x \partial y} \\
\frac{\partial ^2 f}{\partial y \partial x} & \frac{\partial ^2 f}{\partial y^2} 
\end{vmatrix}
= 
\left( \frac{\partial ^2 f}{\partial x^2} \right)_{(x_0,y_0)} \times \left( \frac{\partial ^2 f}{\partial y^2} \right)_{(x_0,y_0)} - 
\left( \frac{\partial ^2 f}{\partial x \partial y} \right)^2_{(x_0,y_0)}
$$

1. If $D > 0$  and $\left( \frac{\partial ^2 f}{\partial x^2} \right)_{(x_0,y_0)} > 0$, then $f$ has a relative minimum at $(x_0,y_0)$.
2. If $D > 0$  and $\left( \frac{\partial ^2 f}{\partial x^2} \right)_{(x_0,y_0)} < 0$, then $f$ has a relative maximum at $(x_0,y_0)$.
3. If $D < 0$ at $(x_0,y_0)$, then f has a saddle point at $(x_0,y_0)$.
4. If $D = 0$, then no conclusion can be drawn at $(x_0,y_0)$.

### Answer Q1i:
The partial differentials
$$ 
\begin{align*}
\frac{\partial f}{\partial x} &= 12x - 3y - 2 \\
\frac{\partial ^2 f}{\partial x^2} &= 12 \\
\frac{\partial ^2 f}{\partial x \partial y} &= -3\\
\frac{\partial f}{\partial y} &= 8y -3x +5 \\
\frac{\partial ^2 f}{\partial y^2} &= 8 \\
\frac{\partial ^2 f}{\partial y \partial x} &= -3 \\
\end{align*}
$$

Stationary points:
$$ 
\begin{align*}
\frac{\partial f}{\partial x} &= 12x - 3y - 2 &= 0 \\  
\frac{\partial f}{\partial y} &= 8y -3x +5 &=0
\end{align*}
$$

Solving for x and y, gives : $ x= \frac{1}{87}, \ y = \frac{-18}{29} $

There is a stationary point at $\left( \frac{1}{87}, \frac{-18}{29} \right)$

Using the theorem of critical points for a function we find $D= (12)(8) - (-3)^2 = 87 > 0$. So $D > 0$ and $\frac{\partial ^2 f}{\partial x^2} = 12 > 0$.  

The function $f(x,y)$ has a relative minimum at $\left( \frac{1}{87}, \frac{-18}{29} \right)$.

### Answer Q1i:
The partial differentials
$$ 
\begin{align*}
\frac{\partial f}{\partial x} &= 2xy - 4y \\
\frac{\partial ^2 f}{\partial x^2} &= 2y \\
\frac{\partial ^2 f}{\partial x \partial y} &= 2x-4\\
\frac{\partial f}{\partial y} &= x^2 -4x -8y +3 \\
\frac{\partial ^2 f}{\partial y^2} &= -8 \\
\frac{\partial ^2 f}{\partial y \partial x} &= 2x -4  \\
\end{align*}
$$

Stationary points:
$$ 
\begin{align*}
\frac{\partial f}{\partial x} &= 2xy -4y &= 0  \quad (1) \\  
\frac{\partial f}{\partial y} &= x^2 -4x -8y +3 &=0 \quad (2)
\end{align*}
$$

Using (1), solving for x and y, gives : $ y=0, \ or \ x = 2 $

Using (2), and $y=0$, then $x=1 \ or \ x=3$

Using (2), and $x=2$, then $y=\frac{-1}{8}$.


There is a stationary points are at $\left( 1, 0 \right)$, $\left( 3, 0 \right)$ and $\left( 2, \frac{-1}{8} \right)$

Using the theorem of critical points for a function we find:
1. At $\left( 1, 0 \right)$, $D=(2*0)(-8) - (2(1) -4)^2 = -4 <0$ and $\frac{\partial ^2 f}{\partial x^2} = 2y = 0$.  The function $f(x,y)$ has a saddle at $\left( 1, 0 \right)$.

2. At $\left( 3, 0 \right)$, $D=(2*0)(-8) - (2(3) -4)^2 = -4 <0$ and $\frac{\partial ^2 f}{\partial x^2} = 2y=0$.  The function $f(x,y)$ has a saddle at $\left( 3, 0 \right)$.

3. At $\left( 2, \frac{-1}{8} \right)$, $D=(2*\frac{-1}{8})(-8) - (2(2) -4)^2 = 1 >0$ and $\frac{\partial ^2 f}{\partial x^2} = 2y = \frac{-2}{8} < 0 $.  The function $f(x,y)$ is a relative maximum $\left( 2, \frac{-1}{8} \right)$.
